```python
import pandas as pd
import numpy as np

df = pd.read_csv('input_day1.txt', sep=' ',
                 header=None,
                 names=['seq1', 'nop1', 'nop2', 'seq2'],
                 index_col=None,
                 usecols=['seq1', 'seq2'],
                 dtype={'seq1': np.int32,
                        'seq2':np.int32})

df_sorted = pd.DataFrame(data={
    "seq1": df['seq1'].sort_values(ascending=True).values,
    "seq2": df['seq2'].sort_values(ascending=True).values})

# calculating distance

df_sorted["diff"] = df_sorted["seq1"] - df_sorted["seq2"]

print("distance", df_sorted['diff'].abs().sum())

# calculating similarity

df_sorted["similar"] = df_sorted["seq1"].apply(lambda x: df_sorted[df_sorted["seq2"] == x].count().iloc[0] * x)

print("similarity", df_sorted['similar'].sum())
```

```python
import math
from typing import List

def get_reports() -> List[int]:
  with open("input_day2.txt") as f:
    lines = f.readlines()
  reports = \
    map(lambda s: [int(e) for e in s],                # convert to int
      map(lambda p: [str.strip(e, '\n') for e in p],  # remove any non numeric characters
        map(lambda x: x.split(' '),                   # split into individual values
          lines)))
  return reports

def is_report_safe(levels: List[int]) -> bool:
  diffs = [levels[i+1] - levels[i] for i in range(len(levels) - 1)]
  if set(diffs) <= {1,2,3} or set(diffs) <= {-1,-2,-3}:
    return True
  return False

# number of safe reports
print(sum(map(is_report_safe, get_reports())))

# number of safe reports with one tolerance of 1
print(\
    sum(
      map(
        lambda r: any([is_report_safe(r[:i] + r[i+1:]) for i in range(len(r))]),
          get_reports())))

```

```python
from posixpath import split
import re

def get_input() -> str:
  with open("input_day3.txt") as f:
    return f.read()

def find_and_multiply(input: str) -> int:
  matches = re.findall("mul\((\d{1,3}),(\d{1,3})\)", input)
  return sum(map(lambda x: int(x[0]) * int(x[1]), matches))

print(find_and_multiply(get_input()))

total = 0
first_run = True
for parts in re.split("do", get_input()):
  if first_run or (parts[0] == '(' and parts[1] == ')'):
    first_run = False
    total += find_and_multiply(parts)

print(total)
```

```python
import re

def findall(pattern, string, rotate=False):
    """Find all two-dimensional matches of the pattern in the string."""
    pattern_grid = []
    for line in pattern.strip().split("\n"):
        pattern_grid.append(list(line))

    grid = []
    for line in string.strip().split("\n"):
        grid.append(list(line))

    if rotate:
        matches = []
        for i in range(4):
            matches.extend([(x,y,i) for (x,y) in _findall(pattern_grid, grid)])
            pattern_grid = _rotate(pattern_grid)

        return matches
    else:
        return _findall(pattern_grid, grid)

def _findall(pattern_grid, grid):
    matches = []
    for x in range(len(grid)):
        for y in range(len(grid[0])):
            match = True
            try:
                for i in range(len(pattern_grid)):
                    for j in range(len(pattern_grid[0])):
                        if pattern_grid[i][j] == ".":
                            continue

                        c = "." if pattern_grid[i][j] == "\\." else pattern_grid[i][j]

                        if grid[x+i][y+j] != c:
                            match = False
                            break

                if match:
                    matches.append((x,y))

            except IndexError:
                pass

    return matches

def _transpose(grid):
    """Transpose the grid."""
    return [[grid[x][y] for x in range(len(grid))] for y in range(len(grid[0]))]

def _flip(grid):
    """Flip the grid horizontally."""
    return [[grid[x][y] for y in reversed(range(len(grid[0])))] for x in range(len(grid))]

def _rotate(grid):
    """Rotate the grid counter-clockwise."""
    return _transpose(_flip(grid))

def get_input() -> str:
  with open("input_day4.txt") as f:
    return f.read()

# part 1
patterns = ["XMAS",
"""X...
.M..
..A.
...S"""]
print(sum(len(findall(pattern, get_input(), rotate=True)) for pattern in patterns))


# part 2
print(
    len(
        findall(
"""M.M
.A.
S.S""", get_input(), rotate=True)))
```

```python
from typing import List, Dict
import math
import itertools

def get_input() -> List[str]:
  with open("input_day5.txt") as f:
    return f.read().split("\n")

def parse_rules(input: List[str]) -> Dict[str, List[str]]:
  rules = {}
  for line in filter(lambda x: "|" in x, input):
    parts = line.split("|")
    if parts[0] not in rules:
      rules[parts[0]] = []
    rules[parts[0]].append(parts[1])
  return rules

def is_reprint_valid(reprint: List[str], rules: Dict[str, List[str]]) -> bool:
  for i in range(len(reprint)):
    for j in range(i+1, len(reprint)):
      if (reprint[i] in rules) and (reprint[j] not in rules[reprint[i]]):
        return False
      if (reprint[j] in rules) and (reprint[i] in rules[reprint[j]]):
        return False
  return True

# part 1
def solve1(input: List[str]):
  result = 0
  rules = parse_rules(input)
  reprints = map(lambda s: s.split(","), filter(lambda x: "," in x, input))
  for reprint in reprints:
    if is_reprint_valid(reprint, rules):
      result += int(reprint[int(len(reprint)/2)])
  return result

print(solve1(get_input()))

# part 2
from functools import cmp_to_key
def solve2(input: List[str]):
  result = 0
  rules = parse_rules(input)
  reprints = map(lambda s: s.split(","), filter(lambda x: "," in x, input))
  for reprint in reprints:
    if not is_reprint_valid(reprint, rules):
      reprint_sorted = sorted(reprint, key=cmp_to_key(lambda x, y: -1 if y in rules[x] else 1))
      result += int(reprint_sorted[int(len(reprint_sorted)/2)])
  return result

print(solve2(get_input()))
```

```python
from typing import List, Tuple, Dict

direction_increment = [
    (-1, 0), # N
    (0, 1),  # E
    (1, 0),  # S
    (0, -1)  # W
]

def get_input() -> List[str]:
  with open("input_day6.txt") as f:
    data = f.read().split("\n")
    for indx in range(len(data)):
      data[indx] = list(data[indx])
    return data[0:len(data)-1]

def find_guard_location(input: List[str]) -> (int, int, int):
  for idx, row in enumerate(input):
    if '^' in row:
      idy = row.index('^')
      return idx, idy, 0
  return -1, -1, -1

def is_out_of_bounds(_input: List[str], _x: int, _y: int):
  return _x < 0 or _x >= len(_input) or _y < 0 or _y >= len(_input[0])

def walk_the_guard(input: List[str], x: int, y: int, direction: int) -> int:
  distance = 1
  seen = {}
  input[x][y] = 'X'
  seen[(x,y)] = [direction]

  while not is_out_of_bounds(input, x, y):
    new_x = x + direction_increment[direction][0]
    new_y = y + direction_increment[direction][1]
    if not is_out_of_bounds(input, new_x, new_y):
      if input[new_x][new_y] == '#':
        direction = (direction + 1) % 4
        continue
      else:
        distance += 1 if input[new_x][new_y] == '.' else 0
        input[new_x][new_y] = 'X'
        if (new_x,new_y) not in seen:
          seen[(new_x,new_y)] = []
        seen[(new_x,new_y)].append(direction)
    x = new_x
    y = new_y

  return distance, seen

def check_for_loop(input, x, y, direction) -> bool:
  visited = {}
  visited[(x,y,direction)] = True
  while not is_out_of_bounds(input, x, y):
    new_x = x + direction_increment[direction][0]
    new_y = y + direction_increment[direction][1]
    if not is_out_of_bounds(input, new_x, new_y):
      if input[new_x][new_y] == '#':
        direction = (direction + 1) % 4
    x = new_x
    y = new_y
    if (x,y,direction) in visited:
      return True
    visited[(x,y,direction)] = True
  return False

def find_new_blockers(input, seen: Dict) -> int:
  count = 0
  for key in seen:
    if len(seen[key]) > 1:
      if (seen[key][0] + 1) % 4 == seen[key][1]:
        new_x = key[0] + direction_increment[seen[key][0]][0]
        new_y = key[1] + direction_increment[seen[key][0]][1]
        if not is_out_of_bounds(input, new_x, new_y) and input[new_x][new_y] != '#':
          count += 1 if check_for_loop(input, new_x, new_y, seen[key][1]) else 0
  return count

input = get_input()
x, y, direction = find_guard_location(input)
distance, seen = walk_the_guard(input, x, y, direction)
print(distance)
count = find_new_blockers(input, seen)
print(count)

# 1719
```

```python
def get_input():
  with open("input_day7.txt") as f:
    return f.readlines()

def dfs(inputs, expected):
  op = ['+', '*']

  if len(inputs) == 2:
    for i in op:
      result = -1
      if i == '+':
        result = inputs[0] + inputs[1]
      elif i == '*':
        result = inputs[0] * inputs[1]
      if result == expected:
        return expected
    return -1
  elif len(inputs) > 2:
    for i in op:
      if i == '+':
        if dfs([inputs[0] + inputs[1], *inputs[2:]], expected) == expected:
          return expected
      elif i == '*':
        if dfs([inputs[0] * inputs[1], *inputs[2:]], expected) == expected:
          return expected

def part1() -> int:
  result = 0
  input = get_input()
  for line in input:
    parts = line.split(':')
    expected_sum = int(parts[0])
    line_parts = list(map(int, parts[1].strip().split(' ')))
    if(dfs(line_parts, expected_sum) == expected_sum):
      result += expected_sum
  return result

def dfs2(inputs, expected):
  op_func = [
      # add
    lambda a,b: a+b,
      # mul
    lambda a,b: a*b,
      # concat
    lambda a,b: int(str(a) + str(b))
  ]
  if len(inputs) == 2:
    for i in range(len(op_func)):
      result = op_func[i](inputs[0], inputs[1])
      if result == expected:
        return expected
  elif len(inputs) > 2:
    for i in range(len(op_func)):
      result = dfs2([op_func[i](inputs[0], inputs[1]), *inputs[2:]], expected)
      if result == expected:
        return expected
  return -1

def part2() -> int:
  result = 0
  input = get_input()
  for line in input:
    parts = line.split(':')
    expected_sum = int(parts[0])
    line_parts = list(map(int, parts[1].strip().split(' ')))
    if(dfs2(line_parts, expected_sum) == expected_sum):
      result += expected_sum
  return result

print(part2())
```

```python
from typing import List, Tuple

def get_input():
  with open("input_day8.txt") as f:
    return f.read().splitlines()

def get_frequencies(input) -> List[str]:
  frequencies = set()
  for line in input:
    frequencies = set(list(frequencies) + list(set(line)))
  frequencies.remove('.')
  return frequencies

def get_antennas(input, signal) -> List[Tuple[int, int]]:
  locations = []
  for idx, line in enumerate(input):
    if signal in line:
      idy = input[idx].index(signal)
      locations.append((idx, idy))
  return locations

def within_bounds(input, x, y) -> bool:
  return x >= 0 and x < len(input) and y >= 0 and y < len(input[0])


def get_distance(loc1, loc2) -> Tuple[int,int]:
  return loc1[0] - loc2[0], loc1[1] - loc2[1]


def get_antinodes(loc1, loc2) -> List[Tuple[int,int]]:
  distance = get_distance(loc1, loc2)
  return [
    (loc1[0] + distance[0], loc1[1] + distance[1]),
    (loc2[0] - distance[0], loc2[1] - distance[1])
  ]

def get_all_antinodes(input, loc1, loc2) -> List[Tuple[int,int]]:
  antinodes = []
  distance = get_distance(loc1, loc2)
  while(within_bounds(input, loc1[0], loc1[1])):
    antinodes.append((loc1[0], loc1[1]))
    loc1 = (loc1[0] + distance[0], loc1[1] + distance[1])
  while(within_bounds(input, loc2[0], loc2[1])):
    antinodes.append((loc2[0], loc2[1]))
    loc2 = (loc2[0] - distance[0], loc2[1] - distance[1])
  return antinodes


input = get_input()
def part1(input):
  antinode_locations = set()
  frequencies = get_frequencies(input)
  for freq in frequencies:
    locations = get_antennas(input, freq)
    for i in range(len(locations)-1):
      for j in range(i+1, len(locations)):
        all_antinodes = get_antinodes(locations[i], locations[j])
        valid_antinodes = list(filter(lambda x: within_bounds(input, x[0], x[1]), all_antinodes))
        for antinode in valid_antinodes:
          antinode_locations.add(antinode)
  return antinode_locations


input = get_input()
def part2(input):
  antinode_locations = set()
  frequencies = get_frequencies(input)
  for freq in frequencies:
    locations = get_antennas(input, freq)
    for i in range(len(locations)-1):
      for j in range(i+1, len(locations)):
        all_antinodes = get_all_antinodes(input, locations[i], locations[j])
        for antinode in all_antinodes:
          antinode_locations.add(antinode)
  return antinode_locations


locations = part2(input)
print(len(locations))
print(locations)
```

```python

def get_input():
  with open("input_day9.txt") as f:
    return f.read().strip()

def expand_disk(input: str) -> str:
  disk = []
  id = 0
  for i in range(0, len(input), 2):
    disk.extend([str(id)] * int(input[i]))
    id += 1
    if i+1 < len(input):
      disk.extend(int(input[i+1]) * '.')
  return disk

def defrag_disk(disk_parts: str) -> str:
  def get_free_block() -> int:
    return disk_parts.index('.') if '.' in disk_parts else 99999999999999

  def swap_block(a, b):
    disk_parts[a], disk_parts[b] = disk_parts[b], disk_parts[a]

  for i in range(len(disk_parts)-1, -1, -1):
    if disk_parts[i] != '.':
      f = get_free_block()
      if i > f:
        swap_block(i, f)
      else:
          break
  return disk_parts

def find_indices(lst, element):
  return [i for i, x in enumerate(lst) if x == element]

def defrag_disk_fullfile(disk_parts) -> str:
  def get_file_length(file_id: str) -> int:
    return len(list(filter(lambda x: x == file_id, disk_parts)))

  def get_start_of_file(file_id: str) -> int:
    return disk_parts.index(file_id)

  def get_free_block(file_len: int) -> int:
    free_blocks = find_indices(disk_parts, '.')
    i = 0
    found_index = free_blocks[0]
    expected_length = file_len - 1
    for i in range(len(free_blocks)):
      free_index = free_blocks[i]
      expected_length = file_len - 1
      j = i
      while(expected_length > 0):
        if j+1 < len(free_blocks) and free_blocks[j+1] == free_blocks[j]+1:
          expected_length -= 1
          j += 1
        else:
          break
      if expected_length == 0:
        return free_index
    return 99999999999999


  def swap_file(a: int, b: int, file_len: int):
    #print(a, b, file_len)
    while file_len > 0 and a < len(disk_parts) and b < len(disk_parts):
      disk_parts[a], disk_parts[b] = disk_parts[b], disk_parts[a]
      a += 1
      b += 1
      file_len -= 1

  for i in range(len(disk_parts)-1, -1, -1):
    if disk_parts[i] != '.':
      file_length = get_file_length(disk_parts[i])
      f = get_free_block(file_length)
      #print(disk_parts[i], file_length, f)
      if i > f:
        swap_file(get_start_of_file(disk_parts[i]), f, file_length)
        #print(disk_parts)
      else:
        i = i - file_length + 1
  return disk_parts


def checksum(disk) -> int:
  hash = 0
  for i in range(len(disk)):
    hash += i * int(disk[i]) if disk[i] != '.' else 0
  return hash

input = get_input()
disk = expand_disk(input)
print(disk)
disk_defrag = defrag_disk_fullfile(disk)
print(disk_defrag)
print(checksum(disk_defrag))
```

```python
def get_input():
  with open("input_day10.txt") as f:
    return f.read().splitlines()

def find_trailheads(map):
  trailheads = []
  for i in range(len(map)):
    for j in range(len(map[i])):
      if map[i][j] == '0':
        trailheads.append((i, j))
  return trailheads

def bfs(map, curr_x, curr_y, summits):
  direction = [(1,0),   #S
               (0,1),   #E
               (-1,0),  #N
               (0,-1)]  #W

  if curr_x < 0 or curr_x >= len(map) or curr_y < 0 or curr_y >= len(map[0]):
    return 0

  if map[curr_x][curr_y] == '9':
    #print(curr_x, curr_y, '->', map[curr_x][curr_y])
    summits.append((curr_x, curr_y))
    return

  for d in direction:
    next_x = curr_x + d[0]
    next_y = curr_y + d[1]
    if next_x < 0 or next_x >= len(map) or next_y < 0 or next_y >= len(map[0]):
      continue
    #print(next_x, next_y)
    if int(map[next_x][next_y]) == int(map[curr_x][curr_y]) + 1:
      bfs(map, next_x, next_y, summits)


map = get_input()
heads = find_trailheads(map)
score = 0
score2 = 0
for head in heads:
  summits = []
  bfs(map, int(head[0]), int(head[1]), summits)
  #print(head, '->', set(summits))
  score += len(set(summits))
  score2 += len(summits)

print(score)
print(score2)
```
